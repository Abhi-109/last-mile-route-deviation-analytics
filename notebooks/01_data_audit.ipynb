## Data Grain
- One row is one stop.
- A route is made of many rows.
- You can group data by route, driver, week, or country.
- Warning: Never average rows directly. You must group them by route first.

## Time Meaning
- Time columns are not clock times (like 2:00 PM).
- They are "minutes from the start."
- Arrived Time: When the driver reached the stop.
- Earliest/Latest Time: The allowed delivery window.
- SLA Failure: This happens if Arrived Time is greater than Latest Time.

## Planned vs. Actual Sequence
- IndexP: The sequence the computer planned.
- IndexA: The sequence the driver actually followed.
- If these numbers do not match, the driver deviated from the path.
- Judge the impact by looking at the whole route, not just one stop.

## Distance Interpretation
- DistanceP/A: The distance from the previous stop only.
- A distance of zero means the driver is at the depot or has two deliveries at one house.
- To find the total distance, sum the rows. Do not average them.
- Compare the total actual distance to the total planned distance to find efficiency loss.
